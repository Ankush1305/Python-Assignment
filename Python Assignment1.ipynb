2.
Create the below pattern using nested for loop in Python.
*
* *
* * *
* * * *
* * * * *
* * * *
* * *
* *
*

```python
a = 5
for i in range(0,a):
    for j in range(0,i+1):
        print("* ",end=" ")
    print()    
for i in range(a,-1,-1):
    for j in range(0,i+1):
        print("* ",end=" ")
    print()
    
```
